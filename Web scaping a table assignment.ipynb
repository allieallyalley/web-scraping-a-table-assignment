```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://webpages.charlotte.edu/mscipion/"

r = requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
table = soup.find("table", class_="table table-sm table-hover screenertable")
print(table)
```

```python
if soup and soup.find("table"):
    table = soup.find("table")
    headers = [th.get_text() for th in table.find_all("th")]
    print(headers)
else:
    print("No table found in the soup object.")
```

```python
titles = [i.text for i in headers if hasattr(i, "text")]
print(titles)
```

```python
data1 = pd.DataFrame(columns=titles)
print(data1)
```
